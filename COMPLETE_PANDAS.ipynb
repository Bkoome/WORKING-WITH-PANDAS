```python
import urllib.request

url = 'https://gist.githubusercontent.com/aakashns/8684589ef4f266116cdce023377fc9c8/raw/99ce3826b2a9d1e6d0bde7e9e559fc8b6e9ac88b/locations.csv'

urllib.request.urlretrieve(url, '.Locations.csv')
```

```python
Locations_pd=pd.read_csv('.Locations.csv')
```

```python
Locations_pd
```

```python
Locations_pd[Locations_pd.location=="Italy"]
```

```python
#creating a column in the covid_italy in the to have a similar column as that of the Location_pd for merging
covid_italy['location'] = "Italy"
```

```python
covid_italy
```

```python
#merging the covid_italy and Locations_pd 
merged_table=covid_italy.merge(Locations_pd, on="location")
```

```python
merged_table
```

```python
merged_table['Cases_in_millions']=merged_table.total_cases * 1e6 / merged_table.population
```

```python
merged_table
```

```python
#writing back to file with added 
pandas_final = merged_table[['date',
                       'new_cases', 
                       'total_cases', 
                       'new_deaths', 
                        
                       'new_tests', 
                       'total_tests', 
                       'Cases_in_millions']]
```

```python
pandas_final
```

```python
pandas_final.to_csv('PANDAS_FINAL.csv', index=False)
```
